# Line plot

A line plot is a type of plot which displays information as a series of data points called 'markers' connected by straight line segments.

```python
x=np.linspace(0,1,5)

y=x+5

plt.plot(x,y,'r:o')

plt.xlabel('X')

plt.ylabel('Y')

plt.title('Line plot')
```

```python
x=np.arange(16)
x
```

# Mean formula: $\mu= \sum_{i=1}^N \frac{X_i}{N}$ 

```python
np.mean(x)
```

# Variance Formula: $\sigma^{2}=\frac{\sum_{i=1}^{N}(X_i-\mu)^{2}}{N}$

```python
np.var(x)
```

```python
np.std(x)
```

# Median Formula: 
$Med(X)=X[\frac{N+1}{2}]$, if N is odd

$Med(X)=\frac{X[\frac{N}{2}]+X[\frac{N}{2}+1]}{2}$, if N is even

```python
np.median(x)
```

# Histogram plot
A histogram plot is an approximate representation of the distribution of numerical data.

```python
plt.hist(x,bins=5,edgecolor='red')

plt.xlabel('X')

plt.ylabel('Frequecy (Count)')

plt.title('Uniform Distribution')
```

```python
# np.random.seed(0)

x=np.random.randn(1000);

plt.hist(x,edgecolor='red');

plt.xlabel('X')

plt.ylabel('Frequency')

plt.title('Standard Normal Distribution')
```

# Data distribution
$$(\mu-1\sigma) \ to \ (\mu+1\sigma)=68.3\%$$
$$(\mu-2\sigma) \ to \ (\mu+2\sigma)=95.5\%$$
$$(\mu-3\sigma) \ to \ (\mu+3\sigma)=99.7\%$$

```python
# np.random.seed(0)

x=np.random.normal(loc=0,scale=2,size=1000);

plt.hist(x,edgecolor='red');

plt.xlabel('X')

plt.ylabel('Frequency')

plt.title('Normal Distribution')
```

```python
plt.boxplot(x);

plt.ylabel('X')

plt.title('Box plot')
```
